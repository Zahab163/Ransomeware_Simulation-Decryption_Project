 ## Key Manager Class

```python
from cryptography.fernet import Fernet

class SecureKeyManager:
  """Secure Key Manager with drive Storage"""
  def __init__(self, save_path):
    self.save_path = save_path
    self.key = None
    self.cipher = None

  def generate_key(self):
    """Generate Secure Encryption Key"""
    self.key = Fernet.generate_key()
    self.cipher = Fernet(self.key)
    print("New encryption key generated")
    return self.key

  def save_key(self, filename="secret_key.json", student_name= "Zahabia Ahmed"):
    """Save Key securely to the drive"""
    if not self.key:
      self.generate_key()
    key_data = {
        'key': self.key.decode(),
        'algorithm': 'Fernet (AES-128)',
        'created_at': datetime.now().isoformat(),
        'student_name': student_name,
        'project': 'Ransomeware Simulation',
        'location': 'Google Drive'

    }
    filepath = os.path.join(self.save_path, filename)
    with open(filepath, 'w') as f:
      json.dump(key_data, f, indent=2)

    print(f"Key saved to: {filepath}")
    print("Key is backed up in Google Drive")
    return filepath

  def load_key(self, filename= 'secret_key.json'):
    """Load Key from Drive"""
    filepath = os.path.join(self.save_path, filename)
    try:
      with open(filepath, 'r') as f:
        key_data = json.load(f)
        self.key = key_data['key'].encode()
        self.cipher = Fernet(self.key)
        print("Key loaded successfully")
        print(f"Created: {key_data.get('created_at', 'Unknown')}")
        print(f"Algorithm: {key_data.get('algorithm', 'Unknown')}")
        print(f"Student Name: {key_data.get('student_name', 'Unknown')}")
        return self.key
    except FileNotFoundError:
      print(f"Key file not found at: {filepath}")
      return None
```

## Generate and Save Your Key

```python
# Set Name
STUDENT_NAME = "Zahabia Ahmed"
```

```python
# Create key manager
key_manager = SecureKeyManager(scripts_folder)
```

```python
# Instantiate the SecureKeyManager
key_save_path = os.path.join(project_folder, 'keys')
os.makedirs(key_save_path, exist_ok=True)
key_manager = SecureKeyManager(save_path=key_save_path)
```

```python
# Save to Drive
key_path = key_manager.save_key(student_name=STUDENT_NAME)
```

```python
# Generate a new key
print("\n--- Generating and Saving Key ---")
key = key_manager.generate_key()
```

```python
## Save to Drive
saved_filepath = key_manager.save_key(filename='my_secret_key.json', student_name=STUDENT_NAME)



print(f"\nKey Manager initialized with save path: {key_save_path}")
print(f"Generated Key (first 10 chars): {key.decode()[:10]}...")
print(f"Key saved to: {saved_filepath}")
```

## Verify Key Works

```python
print("\nTesting encryption with your key...")

cipher = Fernet(key)
test_message = b"Ransomware Simulation Project - Secure Test"
encrypted = cipher.encrypt(test_message)
decrypted = cipher.decrypt(encrypted)

print(f"Original: {test_message}")
print(f"Encrypted: {encrypted[:10]}...")
print(f"Decrypted: {decrypted}")
print("\nEncryption/Decryption working perfectly!")
```

```python
# load the key back to verify
print("\n--- Loading Key for verification ---")
loaded_key = key_manager.load_key(filename='my_secret_key.json')
if loaded_key:
    print(f"Loaded Key (first 10 chars): {loaded_key.decode()[:10]}...")
else:
    print("Failed to load key.")
```
