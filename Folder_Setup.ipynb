## STEP 1: CONNECT TO GOOGLE DRIVE
 This creates a secure connection to your Drive


```python
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

print("Google Drive mounted successfully!")
print("Your Drive is now accessible at: /content/drive/My Drive/")
```


## STEP 2: CREATE SECURE PROJECT FOLDER
Creates a dedicated folder in your Google Drive


```python
import os

# Define your project folder path
project_folder = '/content/drive/My Drive/Ransomware_Simulation_Project'

# Create the folder
os.makedirs(project_folder, exist_ok=True)

print("=" * 60)
print("PROJECT FOLDER CREATED SAFELY")
print("=" * 60)
print(f"Location: {project_folder}")
print(f"All files will be saved here automatically")
print(f"Backed up to your Google Drive")
print("=" * 60)

# List contents to verify
print("\nFolder contents:")
!ls -la "{project_folder}"
```


## STEP 3: CREATE PROJECT STRUCTURE
 Organizes your files in Drive


```python
# Create subdirectories
subfolders = [
    'notebooks',
    'scripts',
    'test_files',
    'encrypted_files',
    'recovered_files',
    'screenshots',
    'reports'
]

for folder in subfolders:
    folder_path = os.path.join(project_folder, folder)
    os.makedirs(folder_path, exist_ok=True)
    print(f"Created: {folder}")

print("\n" + "=" * 60)
print("PROJECT STRUCTURE CREATED")
print("=" * 60)
print("""
ransomware_simulation_project/
├── notebooks/          ← Your Colab notebooks
├── scripts/           ← Python scripts
├── test_files/        ← Original test files
├── encrypted_files/   ← Encrypted files (simulation)
├── recovered_files/   ← Decrypted files
├── screenshots/       ← Screenshots for report
└── reports/           ← Your project reports
""")
```
