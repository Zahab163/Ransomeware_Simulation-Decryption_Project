## Encryption Simulator

```python
class SafeRansomwareSimulator:
    """Safe encryption simulation using Drive"""

    def __init__(self, cipher, source_folder, dest_folder):
        self.cipher = cipher
        self.source = source_folder
        self.dest = dest_folder
        self.encrypted_count = 0

    def encrypt_file(self, source_path, dest_path):
        """Encrypt and save to encrypted folder"""
        try:
            # Read original
            with open(source_path, 'rb') as f:
                data = f.read()

            # Encrypt
            encrypted = self.cipher.encrypt(data)

            # Save to encrypted folder
            with open(dest_path, 'wb') as f:
                f.write(encrypted)

            self.encrypted_count += 1
            return True, dest_path
        except Exception as e:
            return False, str(e)

    def run_simulation(self):
        """Run safe encryption simulation"""
        print("\n" + "=" * 60)
        print("SAFE ENCRYPTION SIMULATION")
        print("=" * 60)
        print(f"Source: {self.source}")
        print(f"Destination: {self.dest}")
        print("Original files remain intact in Drive")
        print("=" * 60)

        encrypted_files = []
        start_time = time.time()

        for filename in os.listdir(self.source):
            source_path = os.path.join(self.source, filename)
            dest_path = os.path.join(self.dest, filename + '.encrypted')

            print(f"\nEncrypting: {filename}")
            success, result = self.encrypt_file(source_path, dest_path)

            if success:
                encrypted_files.append(result)
                print(f" Saved to: encrypted_files/{filename}.encrypted")
            else:
                print(f" Failed: {result}")

            time.sleep(0.1)

        end_time = time.time()

        print("\n" + "=" * 60)
        print("ENCRYPTION COMPLETE")
        print("=" * 60)
        print(f"Encrypted: {self.encrypted_count} files")
        print(f"Time: {end_time - start_time:.2f} seconds")
        print(f"Location: {self.dest}")

        return encrypted_files
```

```python
# Run simulation
simulator = SafeRansomwareSimulator(cipher, test_files_folder, encrypted_folder)
encrypted_files = simulator.run_simulation()
```
